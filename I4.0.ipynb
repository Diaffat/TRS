```python
v = []
for i in range(df.shape[0]):
    c = '0'
    if 6<= int(df["Actual Start Date"][i][12:14])<=13:
        c='S1'
    elif 14<= int(df["Actual Start Date"][i][12:14])<=21:
        c='S2'
    else:
        c = 'S3'
    v.append(c)
len(v)
```

```python
v
```

```python
df = df.assign(Shift = v)
df
```

```python
ad = []
for i in range(df.shape[0]):
    ad.append(df["Actual Start Date"][i][6:10]+"-"+df["Actual Start Date"][i][3:5]+"-"+df["Actual Start Date"][i][0:2])
```

```python
df["Actual Start Date"] = ad
```

```python
df
```

```python
msd = []
for i in range(df.shape[0]):
    c = str(df["Plan  Workstation"][i])
    msd.append(c+" | "+df["Shift"][i]+" | "+df["Actual Start Date"][i])
```

```python
msd
```

```python
df = df.assign(MSD = msd)
df
```

```python
df = df.drop(columns=["Actual Start Date","Actual End Date","Plan  Workstation","Shift"])
df
```

```python
msd2 = []
for i in range(OEE.shape[0]):
    c = str(OEE["Machine"][i])
    msd2.append(c+" | "+OEE["Shift"][i]+" | "+str(OEE["Shift Day"][i])[0:10])
```

```python
msd2
```

```python
OEE = OEE.assign(MSD = msd2)
OEE
```

```python
OEE = OEE.drop(columns=["Shift Day","Quality","Machine","Shift","Meetings  (min)","Machine  Downtime (min)", "Fail Time  (min)","Waste","Breaks  Shift cal. (min)","Good  Parts","Total Time  (min)"])
OEE
```

```python
val = []
for i in range(df.shape[0]):
    if pd.isnull(df["Terminal Left"].loc[i]):
        val.append(0)
    else:
        val.append(1)
```

```python
val2 = []
for i in range(df.shape[0]):
    if pd.isnull(df["Terminal Right"].loc[i]):
        val2.append(0)
    else:
        val2.append(1)
```

```python
df["Terminal Left"] = val
df["Terminal Right"] = val2
```

```python
df
```

```python
df = df.drop(columns=["Machine Setup  Waste (Pcs)","Machine Waste  (Pcs)","Manual Waste  (Pcs)"])
df
```

```python
val3 = []
for i in range(df.shape[0]):
    if pd.isnull(df["Seal Left"].loc[i]):
        val3.append(0)
    else:
        val3.append(1)
val4 = []
for i in range(df.shape[0]):
    if pd.isnull(df["Seal Right"].loc[i]):
        val4.append(0)
    else:
        val4.append(1)

df["Seal Left"] = val3
df["Seal Right"] = val4
```

```python
df
```

```python
df = df.assign(Terminal = df["Terminal Left"] +  df["Terminal Right"])
df = df.assign(Seal = df["Seal Left"] +  df["Seal Right"])
df
```

```python
df = df.drop(columns=["Seal Left","Seal Right","Terminal Left","Terminal Right","Wire Left"])
df
```

```python
df = df.drop(columns=["Batch Size  (Pcs/Bdl)","Good Parts  (Pcs)"])
df
```

```python
df.sort_values('MSD', inplace=True)
df
```

```python
ndf = pd.DataFrame().reindex_like(df)
```

```python
ndf = ndf.dropna()
```

```python
cache = []
for  i in df["MSD"]:
    if i not in cache :
        cf = df[df['MSD'] == i]
        if len(cf) != 0:
            cf['Quantity  (Pcs)'] = cf['Quantity  (Pcs)'].sum()
            cf['Plan Time  (min/Bdl)'] = cf['Plan Time  (min/Bdl)'].sum()
            cf['Prod. Time  (min)'] = cf['Prod. Time  (min)'].sum()
            cf['Setup  Time (min)'] = cf['Setup  Time (min)'].sum()
            cf['Wire Left Length'] = cf['Wire Left Length'].mean()
            cf['Wire Left CrossSection'] = cf['Wire Left CrossSection'].mean()
            cf['Stripping Length Left'] = cf['Stripping Length Left'].mean()
            cf['Stripping Length Right'] = cf['Stripping Length Right'].mean()
            cf['Terminal'] = cf['Terminal'].max()
            cf['Seal'] = cf['Seal'].max()
            cf = cf.drop_duplicates()
            ndf = pd.concat([ndf, cf], axis=0)
            cache.append(i)
        
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import sklearn
from sklearn.svm import SVR
from sklearn.ensemble import RandomForestRegressor
from sklearn.ensemble import GradientBoostingRegressor
from sklearn.neural_network import MLPRegressor
from sklearn.model_selection import train_test_split
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import mean_squared_error, r2_score
```

```python
x_train, x_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 1)
```

# SVR

```python
svr = SVR()

# param = {
#     'C': [i for i in range(1, 3)],
#     'kernel': ['rbf', 'linear', 'poly'],
#     'degree': [i for i in range(2, 4)],}

# grid = GridSearchCV(svr, param, cv=5, scoring='neg_mean_squared_error')
svr.fit(x_train, y_train)
```

```python
ypred = svr.predict(x_test)
rmse_svr = np.sqrt(mean_squared_error(y_test, ypred))
r2_svr = r2_score(y_test, ypred)
print(f"rmse_svr = {rmse_svr}\nr2_svr = {r2_svr}")
```

# RandomForestRegressor

```python
model1 = RandomForestRegressor(n_estimators=400)
model1.fit(x_train, y_train)
ypred = model1.predict(x_test)
rmse_rfr = np.sqrt(mean_squared_error(y_test, ypred))
r2_rfr = r2_score(y_test, ypred)
print(f"rmse_rfr = {rmse_rfr}\nr2_rfr = {r2_rfr}")
```

```python
from matplotlib import pyplot as plt
a = model1.feature_importances_
for i,v in enumerate(a):
    print('Feature: %0d, Score: %.5f'%(i,v))
plt.bar([x for x in range(len(a))], a)
plt.show()
```

# GradientBoostingRegressor

```python
model2 = GradientBoostingRegressor(n_estimators=1000)
model2.fit(x_train, y_train)
ypred = model2.predict(x_test)
rmse_gbr = np.sqrt(mean_squared_error(y_test, ypred))
r2_gbr = r2_score(y_test, ypred)
print(f"rmse_gbr = {rmse_gbr}\nr2_gbr = {r2_gbr}")
```

# MLPRegressor

```python
model3 = MLPRegressor(hidden_layer_sizes=(100,95,85,70,55,40,25,10,))
model3.fit(x_train, y_train)
ypred = model3.predict(x_test)
rmse_mlpr = np.sqrt(mean_squared_error(y_test, ypred))
r2_mlpr = r2_score(y_test, ypred)
print(f"rmse_mlpr = {rmse_mlpr}\nr2_mlpr = {r2_mlpr}")
```
